```python
# Definir fechas
start_date = "2022-01-01"
end_date = "2024-06-30"

# Obtener datos del Ã­ndice
sp500_data = yf.download("^SPX", start=start_date, end=end_date)

# Filtrar por fechas de inicio de trimestre
sp500_data["Month"] = sp500_data.index.month_name()
sp500_data["Quarter"] = sp500_data["Month"].apply(lambda x: int(x[0:1]))

qtr_start_data = sp500_data.groupby(["Year", "Quarter"]).last()

# Exportar a CSV
sp500_data.to_csv("output.csv", index=False)
```
