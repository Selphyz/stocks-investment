```python
def visualize_pe_vs_ps(tickers):
    pe_ratios = []
    ps_ratios = []
    valid_tickers = []

    for ticker_symbol in tickers:
        ticker = Ticker(ticker_symbol)
        info = ticker. key_stats[ticker_symbol]

        # Obtener P/E y P/S ratios
        precio = ticker.key_stats[ticker_symbol]['regularMarketPrice']
        ps_ratio = ticker.key_stats[ticker_symbol]['trailingEps']
        pe_ratio = info.get("forwardPE", None) or info.get("trailingPE", None)

        # Agregar datos válidos a las listas
        if pe_ratio is not None and ps_ratio is not None:
            pe_ratios.append(pe_ratio)
            ps_ratios.append(ps_ratio)
            valid_tickers.append(ticker_symbol)

    # Verificar si hay datos válidos
    if not pe_ratios or not ps_ratios:
        print("No se encontraron suficientes datos para los tickers proporcionados.")
        return

    # Crear scatterplot
    plt.figure(figsize=(10, 6))
    plt.scatter(ps_ratios, pe_ratios, color='blue', alpha=0.7)

    # Agregar etiquetas
    for i, ticker in enumerate(valid_tickers):
        plt.text(ps_ratios[i], pe_ratios[i], ticker, fontsize=9, ha='right')

    plt.title("Relación P/E vs P/S", fontsize=16)
    plt.xlabel("P/S Ratio (Price to Sales)", fontsize=14)
    plt.ylabel("P/E Ratio (Price to Earnings)", fontsize=14)
    plt.grid(alpha=0.4, linestyle='--')
    plt.show()
```

```python
# Función para obtener y graficar ratios P/E y P/S por sector
def visualize_sector_ratios(tickers_by_sector):
    sector_pe = {}
    sector_ps = {}

    for sector, tickers in tickers_by_sector.items():
        pe_values = []
        ps_values = []

        for ticker_symbol in tickers:
            ticker = Ticker(ticker_symbol)
            info = ticker.info

            # Obtener ratios P/E y P/S
            pe_ratio = info.get("forwardPE", None) or info.get("trailingPE", None)
            ps_ratio = info.get("priceToSalesTrailing12Months", None)

            # Agregar valores válidos
            if pe_ratio is not None:
                pe_values.append(pe_ratio)
            if ps_ratio is not None:
                ps_values.append(ps_ratio)

        # Calcular promedio o mediana (usa mediana para evitar sesgos de outliers)
        if pe_values:
            sector_pe[sector] = np.median(pe_values)
        if ps_values:
            sector_ps[sector] = np.median(ps_values)

    # Verificar si tenemos datos
    if not sector_pe or not sector_ps:
        print("No se encontraron suficientes datos para los sectores proporcionados.")
        return

    # Crear scatterplot
    sectors = list(sector_pe.keys())
    pe_ratios = [sector_pe[sector] for sector in sectors]
    ps_ratios = [sector_ps[sector] for sector in sectors]

    plt.figure(figsize=(10, 6))
    plt.scatter(ps_ratios, pe_ratios, color='orange', alpha=0.8)

    # Etiquetas
    for i, sector in enumerate(sectors):
        plt.text(ps_ratios[i], pe_ratios[i], sector, fontsize=9, ha='right')

    plt.title("Relación P/E vs P/S por Sector", fontsize=16)
    plt.xlabel("P/S Ratio (Price to Sales)", fontsize=14)
    plt.ylabel("P/E Ratio (Price to Earnings)", fontsize=14)
    plt.grid(alpha=0.4, linestyle='--')
    plt.show()

```

```python
def calcular_earnings_yield(magic_tickers):
    plt.figure(figsize=(10, 6))

    for ticker in magic_tickers:
        stock = Ticker(ticker)
        history = stock.history(period="5y", interval="3mo")  # Datos históricos de los últimos 5 años, actualizados cada 3 meses
        earnings_yield_data = []

        for date, row in history.iterrows():
            price = row['Close']
            earnings_per_share = stock.info.get('trailingEps')
            if earnings_per_share and price:
                earnings_yield = earnings_per_share / price
                earnings_yield_data.append((date, earnings_yield))

        if not earnings_yield_data:
            print(f"Datos insuficientes para {ticker}.")
            continue

        # Crear DataFrame para facilitar la visualización
        df = pd.DataFrame(earnings_yield_data, columns=['Date', 'Earnings Yield'])
        df.set_index('Date', inplace=True)

        # Agregar datos a la gráfica
        plt.plot(df.index, df['Earnings Yield'], marker='o', linestyle='-', label=ticker)

    # Configuración de la gráfica
    plt.title("Earnings Yield en los últimos 5 años")
    plt.xlabel("Fecha")
    plt.ylabel("Earnings Yield")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def calcular_visualizar_roi_tickers(tickers, labels=None):
    """
    Calcula y visualiza el ROI para un array de tickers con etiquetas.
    
    Args:
        tickers (list): Lista de tickers (e.g., ["AMZN", "AAPL"]).
        labels (list): Lista de etiquetas para cada ticker (opcional).
        
    Returns:
        dict: Diccionario con DataFrames de ROI para cada ticker.
    """
    if labels is None:
        labels = tickers  # Si no se especifican labels, usamos los tickers como etiquetas

    if len(tickers) != len(labels):
        raise ValueError("El número de tickers debe coincidir con el número de etiquetas.")

    roi_data = {}  # Para almacenar los resultados de cada ticker

    plt.figure(figsize=(12, 6))

    for ticker, label in zip(tickers, labels):
        # Descargar precios históricos
        stock = Ticker(ticker)
        precios = stock.history(period="5y", interval="3mo")

        if precios.empty:
            print(f"No se pudieron obtener datos para el ticker {ticker}.")
            continue

        # Seleccionar precios de cierre y calcular ROI
        precios_cierre = precios['Close']
        roi = (precios_cierre / precios_cierre.shift(1) - 1) * 100

        # Guardar los datos en un DataFrame
        datos_roi = pd.DataFrame({
            'Fecha': precios_cierre.index,
            'Precio Cierre': precios_cierre,
            'ROI (%)': roi
        }).dropna()

        roi_data[ticker] = datos_roi  # Guardar datos para este ticker

        # Graficar ROI
        plt.plot(datos_roi['Fecha'], datos_roi['ROI (%)'], marker='o', linestyle='-', label=label)

    # Configuración del gráfico
    plt.title("Return on Investment (ROI) - Últimos 5 años")
    plt.xlabel("Fecha")
    plt.ylabel("ROI (%)")
    plt.axhline(0, color='gray', linestyle='--', linewidth=1)  # Línea en ROI = 0
    plt.legend()  # Mostrar leyenda
    plt.grid()
    plt.show()

    return roi_data
```

```python
# Funcion que recibe un array de tickers y visualiza el precio de cierre en los últimos 5 años
def visualizar_precio_cierre(tickers):
    """
    Visualiza el precio de cierre de los tickers proporcionados en los últimos 5 años.

    Args:
        tickers (list): Lista de tickers (e.g., ["AMZN", "AAPL"]).
    """
    plt.figure(figsize=(12, 6))

    for ticker in tickers:
        stock = Ticker(ticker)
        history = stock.history(period="5y", interval="1d")  # Datos históricos de los últimos 5 años

        if history.empty:
            print(f"No se pudieron obtener datos para el ticker {ticker}.")
            continue

        # Graficar el precio de cierre
        plt.plot(history.index, history['Close'], label=ticker)

    # Configuración del gráfico
    plt.title("Precio de Cierre en los últimos 5 años")
    plt.xlabel("Fecha")
    plt.ylabel("Precio de Cierre")
    plt.legend()
    plt.grid()
    plt.show()
```

```python
def obtener_datos_financieros_alpha_vantage(tickers, api_key):
    """
    Obtiene estados financieros (ingresos y balances) de múltiples tickers usando Alpha Vantage.
    Incluye un retraso para respetar los límites del plan gratuito.

    Args:
        tickers (list): Lista de tickers (e.g., ["AAPL", "AMZN"]).
        api_key (str): Clave de la API de Alpha Vantage.

    Returns:
        dict: Diccionario con datos financieros por ticker.
    """
    base_url = "https://www.alphavantage.co/query"
    datos_financieros = {}

    for i, ticker in enumerate(tickers):
        print(f"Obteniendo datos para {ticker} ({i+1}/{len(tickers)})...")

        try:
            # Solicitar estado de resultados
            url_income = f"{base_url}?function=INCOME_STATEMENT&symbol={ticker}&apikey={api_key}"
            response_income = requests.get(url_income)
            income_data = response_income.json()

            # Solicitar balance general
            url_balance = f"{base_url}?function=BALANCE_SHEET&symbol={ticker}&apikey={api_key}"
            response_balance = requests.get(url_balance)
            balance_data = response_balance.json()

            # Guardar datos en el diccionario
            datos_financieros[ticker] = {
                "income_statement": income_data.get("quarterlyReports", []),
                "balance_sheet": balance_data.get("quarterlyReports", [])
            }

            # Agregar un retraso para respetar los límites
            if i < len(tickers) - 1:  # Evitar retraso innecesario después de la última solicitud
                time.sleep(12)  # 12 segundos entre solicitudes

        except Exception as e:
            print(f"Error al obtener datos para {ticker}: {e}")

    return datos_financieros
```

```python
def calcular_roc_alpha_vantage(datos_financieros):
    """
    Calcula el ROC para múltiples tickers usando datos obtenidos de Alpha Vantage.

    Args:
        datos_financieros (dict): Diccionario con los datos financieros obtenidos de Alpha Vantage.

    Returns:
        dict: Diccionario con DataFrames que contienen las fechas y el ROC calculado para cada ticker.
    """
    rocs = {}

    for ticker, datos in datos_financieros.items():
        fechas = []
        rocs_ticker = []

        # Extraer datos de ingresos y balances
        income_reports = datos["income_statement"]
        balance_reports = datos["balance_sheet"]

        for income, balance in zip(income_reports, balance_reports):
            try:
                # Calcular EBIT (Operating Income) y Capital Empleado
                ebit = float(income["operatingIncome"])
                total_assets = float(balance["totalAssets"])
                current_liabilities = float(balance["totalCurrentLiabilities"])
                capital_empleado = total_assets - current_liabilities

                # Calcular ROC
                roc = (ebit / capital_empleado) * 100
                fechas.append(income["fiscalDateEnding"])
                rocs_ticker.append(roc)
            except (KeyError, ValueError):
                # Si faltan datos o hay problemas con conversiones, ignoramos ese trimestre
                continue

        # Crear un DataFrame para este ticker
        rocs[ticker] = pd.DataFrame({"Fecha": fechas, "ROC (%)": rocs_ticker})

    return rocs
```

```python
def visualizar_roc(rocs):
    """
    Grafica el ROC para múltiples tickers.

    Args:
        rocs (dict): Diccionario con DataFrames de ROC calculado para cada ticker.
    """
    plt.figure(figsize=(12, 6))

    for ticker, df in rocs.items():
        # Asegurar que las fechas están ordenadas
        df["Fecha"] = pd.to_datetime(df["Fecha"])
        df = df.sort_values(by="Fecha")

        # Graficar el ROC
        plt.plot(df["Fecha"], df["ROC (%)"], marker="o", linestyle="-", label=ticker)

    # Configurar el gráfico
    plt.title("Return on Capital (ROC) de Tickers - Historico")
    plt.xlabel("Fecha")
    plt.ylabel("ROC (%)")
    plt.axhline(0, color="gray", linestyle="--", linewidth=1)  # Línea en ROC = 0
    plt.legend()
    plt.grid()
    plt.show()
```
